```python
# We're checking if the actual weekly operating cost is more than what the model says (₹4000).
#
# Hypotheses:
# H₀ (null): Weekly cost is ₹4000 → model is correct
# H₁ (alternative): Weekly cost is more than ₹4000 → owners say it's higher
#
# This is a one-tailed test because we're only checking if cost is higher, not lower.
# We're using a 5% significance level (α = 0.05), which means we're okay with a 5% chance of being wrong if we reject H₀.

```

```python
import math

# Sample mean
x_bar = 3050

# Theoretical mean cost from model: ₹1000 + ₹5×600
mu = 4000

# Standard deviation of cost: ₹5 × 25
sigma = 125

# Sample size
n = 25
```

```python

t = (x_bar - mu) / (sigma / math.sqrt(n))
print("Test statistic (t):", round(t, 2))
```

```python
# For α = 0.05 (one-tailed), Z-critical = 1.645
z_critical = 1.645
print("Critical value (Z):", z_critical)

# Compare t with Z-critical
if t > z_critical:
    print("Reject H₀ → owners are right, cost is higher")
else:
    print("Fail to reject H₀ → model stands, no proof cost is higher")
```

```python
#The sample mean ₹3050 is **much lower** than the model's ₹4000.

#So, we **fail to reject** the null hypothesis.
#There is **no statistical evidence** that the actual weekly cost is higher than the model.
#In fact, the model may be **overestimating** the real cost.
```
