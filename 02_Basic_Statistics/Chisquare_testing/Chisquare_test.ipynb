```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
#data from the assignment
data = {
    'Satisfaction': ['Very Satisfied', 'Satisfied', 'Neutral', 'Unsatisfied', 'Very Unsatisfied'],
    'Smart_Thermostat': [50, 80, 60, 30, 20],
    'Smart_Light': [70, 100, 90, 50, 50]
}

df = pd.DataFrame(data)
df.set_index('Satisfaction', inplace=True)
df
```

```python
# Running the test
chi2, p, dof, expected = chi2_contingency(df)
```

```python
print("Chi-Square Value:", round(chi2, 4))
print("Degrees of Freedom:", dof)
print("P-Value:", round(p, 4))
```

```python
# Expected values calculated by formula
expected_df = pd.DataFrame(expected, index=df.index, columns=df.columns)
expected_df
```

```python
# Critical value for df=4, alpha=0.05
critical_value = 9.488

if chi2 > critical_value:
    print("Conclusion: Reject H0 → Device type affects satisfaction")
else:
    print("Conclusion: Fail to reject H0 → No strong link between device and satisfaction")
```
