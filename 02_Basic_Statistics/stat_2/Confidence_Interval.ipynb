```python
#importing required libraries
import numpy as np
from scipy.stats import t
from scipy.stats import norm

# Sample data
durability = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]

#Basic calculations
n = len(durability)
mean = np.mean(durability)
std_sample = np.std(durability, ddof=1)


print("Sample Size:", n)
print("Sample Mean:", round(mean, 4))
print("Sample Std:", round(std_sample, 4))

```

```python
# Confidence Interval using t-distribution
confidence = 0.99
df = n - 1
t_value = t.ppf((1 + confidence) / 2, df)
margin_t = t_value * (std_sample / np.sqrt(n))
lower_t = mean - margin_t
upper_t = mean + margin_t


print("99% Confidence Interval using t-distribution:")
print("Lower Bound:", round(lower_t, 4))
print("Upper Bound:", round(upper_t, 4))
```

```python
# Confidence Interval using z-distribution
std_population = 0.2
z_value = norm.ppf((1 + confidence) / 2)
margin_z = z_value * (std_population / np.sqrt(n))
lower_z = mean - margin_z
upper_z = mean + margin_z


print("99% Confidence Interval using z-distribution:")
print("Lower Bound:", round(lower_z, 4))
print("Upper Bound:", round(upper_z, 4))
```

```python
# Summary of both intervals
print("\n--- Summary ---")
print("Sample Mean:", round(mean, 4))
print("t-distribution interval : (", round(lower_t, 4), ",", round(upper_t, 4), ")")
print("z-distribution interval : (", round(lower_z, 4), ",", round(upper_z, 4), ")")
```
