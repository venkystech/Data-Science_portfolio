```python
#program to check the givem]n number if prime or not
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
else:
    print("Not a prime number")
```

```python
#giving product of two nubers
import random
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Asking the user to enter the product
print(f"What is {num1} Ã— {num2} ?")
user_answer = int(input("Your answer: "))

correct_answer = num1 * num2

if user_answer == correct_answer:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The right answer is {correct_answer}.")
```

```python
# print all even or odd number from the range of 100,200
for i in range(100, 201):
    square = i ** 2
    if i % 2 == 0:
        print(f"{i} is even and its square is {square}")
    else:
        print(f"{i} is odd and its square is {square}")
```

```python
#program to check no.of words in a sentence
sentence=input("enter a sentence:")
words=sentence.split()
print(f"the number of words in the sentence is:{words}")
#for counting frequesny of each word
for word in words:
    count = 0
    for w in words:
        if word == w:
            count += 1
    print(word, "appeared", count, "times")
```

```python
#checking palindrome of a given string
string = input("Enter a string: ")

if string == string[::-1]:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```
